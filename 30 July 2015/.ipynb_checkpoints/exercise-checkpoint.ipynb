```python
# with katie rose pipkin
beginning = ["isotomic", "lobed", "perianth", "terrestrial", "toothed", "trapeziform", "tree", "urn-shaped", "variegated", "veined", "whorled", "iridescent"]
middle = ["alate", "ambrosia", "metathoracic", "proboscis", "sporophagic", "villose", "hypypharynx", "hyaline", "hemophagic", "acariphagous"]
end = ["azimuth", "color index", "constellation", "ecliptic", "albedo", "metallicity", "opacity", "periapsis", "transit", "zenith"]

import random

our_beginning = random.choice(beginning)
our_middle = random.choice(middle)
our_end = random.choice(end)

final= [our_beginning, our_middle, our_end]
random.shuffle(final)

final = ' '.join(final)

print final
```

```python
for b in beginning:
    for m in middle:
        for e in end:
            print b + " " + m + " " + " " + e
```
